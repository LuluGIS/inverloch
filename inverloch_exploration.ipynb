```python
# Function to create marine habitat map

def biotope_polygon_designer(habitat, featureGroup, key, value):
    habitatPoly = folium.GeoJson(data=habitat[(habitat['HAB3BIOTTL'] == key)],
                            style_function=lambda x: {'fillColor': value, 'color': value, 'weight': 1,'fillOpacity': 0.7},
                            tooltip=folium.GeoJsonTooltip(
                            fields=['BIOTC_CDE', 'BIOT_LBL', 'BIOT_LVL', 'HAB3BIOTTL'],
                            aliases=['Biotope code: ', 'Biotope label: ', 'Biotope level: ', 'Biotope habitat title: '],
                            localize=True)).add_to(featureGroup)
    return featureGroup




def inverloch_marine_map(marineMap, bathyContours, zoiInverloch, habitatDict):

    # Define location map and add basemaps as first layers
    loc_map = folium.Map(location = [-38.68898706877174, 145.72004848300898], zoom_start=11, tiles=None)
    # base maps
    folium.TileLayer(name='Open Street Map Standard', 
                     overlay=True, 
                     control=False, 
                     tiles='OpenStreetMap').add_to(loc_map)
    folium.TileLayer(name='Satellite overview', 
                     overlay=True, 
                     tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                     attr='Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community').add_to(loc_map)
    
    # Add zone of interest boundaries
    zoneOfInterest = folium.FeatureGroup(name="Zone of interest", overlay=True)
    folium.GeoJson(zoiInverloch, 
                   style_function=lambda x:{'fillColor': False, 'color': '#ff0000', 'fillOpacity': 0.01, "dashArray": "5, 5",},
                   tooltip=folium.GeoJsonTooltip(fields=['ZOI_INFO'],
                                                 aliases=['Zone of interest info: '],
                                                 localize=True)).add_to(zoneOfInterest)
    zoneOfInterest.add_to(loc_map)
    
    for key, value in habitatDict.items():

        habitatPoly = biotope_polygon_designer(habitat=marineMap, 
                                               featureGroup=folium.FeatureGroup(name=key, overlay=True),
                                               key=key, value=value)
        # Add separate groups to map
        habitatPoly.add_to(loc_map)
    
    # Add bathymetric contours to indicate sea depth levels
    bathyCont = folium.FeatureGroup(name="Bathymetric contours", overlay=True)
    folium.GeoJson(bathyContours, 
                   style_function=lambda x:{'fillColor': '#ffffff', 'color': '#ffffff',
                   'weight': 2,
                   'fillOpacity':0.8},
                   tooltip=folium.GeoJsonTooltip(fields=['CONTOUR'],
                                                 aliases=['Depth: '],
                                                 localize=True)).add_to(bathyCont)
    bathyCont.add_to(loc_map)


    folium.LayerControl(collapsed=True).add_to(loc_map)

    # Build a dragable map-legend
    template = """
    {% macro html(this, kwargs) %}

    <!doctype html>
    <html lang="en">
    <head>
    <meta charset="utf-8">
    <meta name="viewport" content="width=device-width, initial-scale=1">
    <title>Inverloch's marine biotope habitats</title>
    <link rel="stylesheet" href="//code.jquery.com/ui/1.12.1/themes/base/jquery-ui.css">

    <script src="https://code.jquery.com/jquery-1.12.4.js"></script>
    <script src="https://code.jquery.com/ui/1.12.1/jquery-ui.js"></script>
    
    <script>
    $( function() {
        $( "#maplegend" ).draggable({
                        start: function (event, ui) {
                            $(this).css({
                                right: "auto",
                                top: "auto",
                                bottom: "auto"
                            });
                        }
                    });
    });

    </script>
    </head>
    <body>

    
    <div id='maplegend' class='maplegend' 
        style='position: absolute; z-index:9999; border:2px solid grey; background-color:rgba(255, 255, 255, 0.8);
        border-radius:6px; padding: 10px; font-size:14px; right: 20px; bottom: 20px;'>

    <div class='legend-title'>Marine biotope habitats</div>
    <div class='legend-scale'>
    <ul class='legend-labels'> 
        <li><span style='background:#c644c8;opacity:0.7;'></span>High energy infralittoral rock</li>
        <li><span style='background:#d783aa;opacity:0.7;'></span>Littoral sand</li>
        <li><span style='background:#86e3cf;opacity:0.7;'></span>Sublittoral rhodolith beds</li>
        <li><span style='background:#4553d1;opacity:0.7;'></span>Sublittoral sand and muddy sand</li>
        <li><span style='background:#dca448;opacity:0.7;'></span>High energy littoral rock</li>
        <li><span style='background:#793bc9;opacity:0.7;'></span>Saltmarsh and reedbeds</li>
        <li><span style='background:#000000;opacity:0.7;'></span>Uncategoried</li>
        <li><span style='background:#d6817b;opacity:0.7;'></span>High energy open-coast circalittoral rockr</li>
        <li><span style='background:#f3f01d;opacity:0.7;'></span>Mangrove</li>
        <li><span style='background:#259ddd;opacity:0.7;'></span>Sublittoral seagrass beds</li>
        <li><span style='background:#70d188;opacity:0.7;'></span>Moderate energy infralittoral rock</li>
    </ul>
    <div class='legend-title'>Other</div>
    <ul class='legend-labels'>
        <li><span style='background:#ffffff;opacity:0.7;'></span>Bathymetric contours</li>
        <li><span style='background:#ff0000;opacity:0.7;'></span>Zone of interest</li>
    </ul> 

    </div>
    
    
    
    </body>
    </html>

    <style type='text/css'>
    .maplegend .legend-title {
        text-align: left;
        margin-bottom: 5px;
        font-weight: bold;
        font-size: 90%;
        }
    .maplegend .legend-scale ul {
        margin: 0;
        margin-bottom: 5px;
        padding: 0;
        float: left;
        list-style: none;
        }
    .maplegend .legend-scale ul li {
        font-size: 80%;
        list-style: none;
        margin-left: 0;
        line-height: 18px;
        margin-bottom: 2px;
        }
    .maplegend ul.legend-labels li span {
        display: block;
        float: left;
        height: 16px;
        width: 30px;
        margin-right: 5px;
        margin-left: 0;
        border: 1px solid #999;
        }
    .maplegend .legend-source {
        font-size: 80%;
        color: #777;
        clear: both;
        }
    .maplegend a {
        color: #777;
        }
    </style>
    {% endmacro %}"""

    macro = MacroElement()
    macro._template = Template(template)

    loc_map.get_root().add_child(macro)

    return loc_map

#test = inverloch_marine_map(marineMap=marineMap, bathyContours=bathyContours, zoiInverloch=zoiInverloch, habitatDict=biotopeTtl_dict)

#test


inverloch_marine_map(marineMap=marineMap, bathyContours=bathyContours, zoiInverloch=zoiInverloch, habitatDict=biotopeTtl_dict).save('inverloch-marine-biotope-case-study-map.html')
```
